```python
import time

import numpy as np

import pandas
import sqlite3
import ciso8601
import talib
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

def get_df(path=None):
    if path is None:
        path = ''

    conn = sqlite3.connect(path + "bitmex.db")
    c = conn.cursor()

    s = time.perf_counter()
    symbol = 'XBTUSD'
    count = 100000
    c.execute("SELECT * FROM tradebin1m"
              " ORDER BY timestamp"
              " LIMIT ?"
              " OFFSET (SELECT COUNT(*) FROM tradebin1m) - ?"
              , [count, count]
              )
    print('execute : ', (time.perf_counter() - s) / 1_000_000)

    s = time.perf_counter()
    d = c.fetchall()
    print('fetchall : ', (time.perf_counter() - s) / 1_000_000)

    close = [item[6] for item in d]
    
    conn.close()
    return numpy.array(close)

BS_PATH = '/Users/bsdo/projects/Python/BSTrade/BSTrade/Data/bitmex/'
close = get_df(BS_PATH)
```

```python

def get_bbr(r = range(20,5000,20)):
    global close
    a = []
    for minute in r:
        period = minute
        upper, middle, lower = talib.BBANDS(close, timeperiod=period)
        a.append((close - lower) / (upper - lower))
    return numpy.array(a)
```
