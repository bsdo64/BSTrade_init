```python
import time

import numpy

import pandas
import sqlite3
import ciso8601

conn = sqlite3.connect("bitmex.db")
c = conn.cursor()

s = time.perf_counter()
symbol = 'XBTUSD'
count = 100000
c.execute("SELECT * FROM tradebin1m"
          " ORDER BY timestamp"
          " LIMIT ?"
          " OFFSET (SELECT COUNT(*) FROM tradebin1m) - ?"
          , [count, count]
          )
print('execute : ', (time.perf_counter() - s) / 1000_000)

s = time.perf_counter()
d = c.fetchall()
print('fetchall : ', (time.perf_counter() - s) / 1000_000)

conn.close()
```

```python
ts = [ciso8601.parse_datetime(item[1]) for item in d]
close = [item[6] for item in d]
```

```python
def get_ratio(before, after):
    return 1 + (after - before) / before

get_ratio(100, 25)
```

```python
money = 100

order = []
for i in range(len(df.close)):
    _rsi = rsi['10'][i]
    price = df.close[i]
    
    if pandas.isna(_rsi):
        order.append((money, None, None, None, None, None))
        continue
    
    last_order = order[-1]
    last_money = last_order[0]
    last_pos = last_order[1]
    last_price = df.close[i-1]
    
    ratio = get_ratio(last_price, price)
    
    if _rsi < 20:
        
        if last_pos == None:
            order.append((last_money, 'buy', _rsi, rsi.loc[i].idxmax(), rsi.loc[i].idxmin(), 'buy_sig'))
        else:
            order.append((last_money * ratio, last_pos, _rsi, rsi.loc[i].idxmax(), rsi.loc[i].idxmin(), None))
        
    elif _rsi > 90:
        
        if last_pos == 'buy':
            order.append((last_money * ratio, 'sell', _rsi, rsi.loc[i].idxmax(), rsi.loc[i].idxmin(), 'sell_sig'))
        else:
            order.append((last_money, None, _rsi, rsi.loc[i].idxmax(), rsi.loc[i].idxmin(), None))
    
    else:
        if last_pos == 'buy':
            order.append((last_money * ratio, 'buy', _rsi, rsi.loc[i].idxmax(), rsi.loc[i].idxmin(), None))
        else:
            order.append((last_money, None, _rsi, rsi.loc[i].idxmax(), rsi.loc[i].idxmin(), None))
```

```python
money = [tup[0] for tup in order]
pos = [tup[1] for tup in order]
r = [tup[2] for tup in order]
r_max = [tup[3] for tup in order]
r_min = [tup[4] for tup in order]
sig = [tup[5] for tup in order]

len(money), len(pos), len(r)
backt = pandas.DataFrame({'money': money, 'pos': pos, 'r': r, 'r_max': r_max, 'r_min': r_min, 'sig': sig})
```
